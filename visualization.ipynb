```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the CSV file into a DataFrame
data = pd.read_csv('./results.csv')

# Specify the column names for x and y axes
x_column = data.columns[0]
y_column = data.columns[1]

# Create a larger figure
plt.figure(figsize=(10, 6))  # Adjust the width and height as desired

# Plot the histogram
plt.hist(data[x_column], bins='auto', alpha=0.7)  # Adjust the number of bins as needed
plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.title('Histogram')

# Show the plot
plt.show()
```
